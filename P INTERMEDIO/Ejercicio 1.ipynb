```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
#Consideremos tres puntos en el plano que formen un triángulo equilátero.

A = [0, 1, 0.5]
B = [0, 0, 2]

plt.plot(A, B, 'bo')
```

```python
#Elegimos un punto aleatorio para considerarlo nuestra posición actual.

D1 = np.random.rand()
D2 = (1 - D1)

plt.plot(A, B, 'bo')
plt.plot(D1, D2, 'ro')

D1 = [D1]
D2 = [D2]
```

```python
x = 0
n = 10
l = [0, 1, 2]

while x < n:
    i = random.choice(l)
    D1.extend([(D1[-1] + A[i]) / 2])
    D2.extend([(D2[-1] + B[i]) / 2])
    x += 1
    
plt.plot(A, B, 'bo')
plt.plot(D1, D2, 'ro')
```

```python
#Implementamos una función.

def Sierpinsky(): 
    A = [0, 1, 0.5]
    B = [0, 0, 2]
    
    D1 = np.random.rand()
    D2 = (1 - D1)
    D1 = [D1]
    D2 = [D2]
    
    plt.plot(A, B, 'bo')
    x = 0
    l = [0,1,2]
    
    while x < n:
        i = random.choice(l)
        
        D1.extend([(D1[-1] + A[i]) / 2])
        D2.extend([(D2[-1] + B[i]) / 2])
        
        x += 1
        
    plt.plot(D1, D2, 'ro')
```
