#### Implementación de un código para obtener una aproximación del triángulo de Sierpinski.

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
#Consideremos tres puntos en el plano que formen un triángulo equilátero.

a = [0, 1, 0.5]
b = [0, 0, 2]

plt.plot(a, b, 'mo')
```

```python
#Elegimos un punto aleatorio para considerarlo nuestra posición actual.

t1 = np.random.rand()
t2 = (1 - t1)

plt.plot(a, b, 'mo')
plt.plot(t1, t2, 'ko')

t1 = [t1]
t2 = [t2]
```

```python
x = 0
n = 10
l = [0, 1, 2]

while x < n:
    i = random.choice(l)
    t1.extend([(t1[-1] + a[i]) / 2])
    t2.extend([(t2[-1] + b[i]) / 2])
    x += 1
    
plt.plot(a, b, 'mo')
plt.plot(t1, t2, 'ko')
```

```python
#Implementamos una función (se usa "def" al estar usando el kernel de python).

def Sierpinski(n):
    a = [0, 1, 0.5]
    b = [0, 0, 2]
    t1 = np.random.rand()
    t2 = (1 - t1)
    t1 = [t1]
    t2 = [t2]
    plt.plot(a, b, 'mo')
    x = 0
    l = [0, 1, 2]
    
    while x < n:
        i = random.choice(l)
        t1.extend([(t1[-1] + a[i]) / 2])
        t2.extend([(t2[-1] + b[i]) / 2])
        x += 1
        
    plt.plot(t1, t2, 'ko')
```

Jugamos con el número de puntos:
